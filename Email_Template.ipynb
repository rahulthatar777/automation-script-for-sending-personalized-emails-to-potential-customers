```python



# Function to send a personalized email
def send_email(receiver_email, store_name, subject, html_content):
    try:
        # Customize HTML content with store name
        personalized_content = html_content.replace("{store_name}", store_name)

        # Set up the email message
        msg = MIMEMultipart("alternative")
        msg["From"] = SENDER_EMAIL
        msg["To"] = receiver_email
        msg["Subject"] = subject

        # Attach the HTML content
        msg.attach(MIMEText(personalized_content, "html"))

        # Connect to the SMTP server
        server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
        server.starttls()
        server.login(SENDER_EMAIL, SENDER_PASSWORD)

        # Send the email
        server.sendmail(SENDER_EMAIL, receiver_email, msg.as_string())
        print(f"Email successfully sent to {receiver_email} ({store_name})")

        # Close the connection
        server.quit()
    except Exception as e:
        print(f"Failed to send email to {receiver_email}: {e}")
```
